##DICA 5

###Rodando o Google Colab Localmente.

Fazer em sua máquina local:

**1)   Instalar o Jupyter** (Sugestão instalar o Jupyter pelo Anaconda Toolkit)

https://www.anaconda.com/products/individual

**2) Instalar e Habilitar a extensão `jupyter_http_over_ws`**

`pip install jupyter_http_over_ws`

`jupyter serverextension enable --py jupyter_http_over_ws`

**3) Rodar o Jupyter e permitir acesso ao Google Colab**

`jupyter notebook --NotebookApp.allow_origin='https://colab.research.google.com' --port=8888 --NotebookApp.port_retries=0`

**4) Conectar a uma RunTime Local**

No Colab, clique no botão "Connect" e selecione "Connect to local runtime" Digite o URL da etapa anterior na caixa de diálogo exibida e clique no botão "Connect".
